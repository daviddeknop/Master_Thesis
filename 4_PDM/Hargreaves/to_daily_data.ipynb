```python
import pandas as pd

# Specify the path to your pickle file
file_path = r'C:\Users\david\Documents\master_thesis\data\Zwalm_data\preprocess_output\zwalm_p_thiessen.pkl'

# Load the pickle file into a DataFrame
data = pd.read_pickle(file_path)

# Ensure Timestamp is in datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Sum the values of P_thiessen columns over each day
data_daily = (
    data.groupby(data['Timestamp'].dt.date)  # Group by the date part of the timestamp
    .sum(numeric_only=True)  # Sum only numeric columns
    .reset_index()  # Reset index to make 'date' a column
)

# Rename the 'date' column to 'Timestamp'
data_daily.rename(columns={'Timestamp': 'Timestamp'}, inplace=True)

# Keep only the 'Timestamp' and 'P_thiessen' columns
data_daily = data_daily[['Timestamp', 'P_thiessen']]

# Save the filtered DataFrame to a new pickle file named 'p_thiessen_daily.pkl'
daily_file_path = r'C:\Users\david\Documents\master_thesis\data\Zwalm_data\preprocess_output\p_thiessen_daily.pkl'
data_daily.to_pickle(daily_file_path)

# Show the first few rows of the resulting DataFrame
print(data_daily.head())
print(f"Daily resolution data with only 'Timestamp' and 'P_thiessen' has been saved to {daily_file_path}")

```

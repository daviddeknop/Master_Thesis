```python
import pandas as pd
from pathlib import Path


# Construct the path to the original pickle file
pickle_file_path = current_dir / "data" / "Zwalm_data" / "preprocess_output" / "forcings_ERA5.pkl"

# Check if the pickle file exists before proceeding
if pickle_file_path.exists():
    # Read the pickle file into a DataFrame
    data = pd.read_pickle(pickle_file_path)

    # Extract the required columns using .loc to avoid chained indexing
    data_filtered = data.loc[:, ['date', 'potential_evaporation_sum', 'total_precipitation_sum']]

    # Convert 'date' column to datetime (Timestamp) using direct column assignment
    data_filtered['date'] = pd.to_datetime(data_filtered['date'])

    # Rename the 'date' column to 'Timestamp'
    data_filtered.rename(columns={'date': 'Timestamp'}, inplace=True)

    # Filter rows from 1972-07-01 onward
    data_filtered = data_filtered[data_filtered['Timestamp'] >= '1972-07-01']

    # Reset the index to avoid saving the original index
    data_filtered.reset_index(drop=True, inplace=True)

    # Construct the path for the new pickle file
    new_pickle_file_path = current_dir / "data" / "Zwalm_data" / "preprocess_output" / "forcings_ERA5_david.pkl"

    # Save the filtered DataFrame to the new pickle file
    data_filtered.to_pickle(new_pickle_file_path)

    print(f"New pickle file has been saved at {new_pickle_file_path}")
else:
    print(f"Error: The file {pickle_file_path} does not exist.")

```

```python
import pandas as pd
from pathlib import Path


# Construct the path to the input pickle file
input_pickle_file_path = current_dir / "data" / "Zwalm_data" / "preprocess_output" / "forcings_ERA5_david.pkl"

# Check if the input pickle file exists before proceeding
if input_pickle_file_path.exists():
    # Read the input pickle file into a DataFrame
    data = pd.read_pickle(input_pickle_file_path)

    # Expand the DataFrame to include 24 hours for each day
    hourly_data = data.loc[data.index.repeat(24)].reset_index(drop=True)

    # Generate hourly timestamps in the correct format
    hourly_offsets = list(range(24)) * len(data)  # Create a list of 0-23 repeated for each day
    hourly_data['Timestamp'] = (
        pd.to_datetime(data['Timestamp'].repeat(24)).reset_index(drop=True)  # Repeat each day 24 times
        + pd.to_timedelta(hourly_offsets, unit='H')  # Add hours 0 to 23 for each day
    )

    # Divide the daily values by 24 to distribute evenly across the hours
    hourly_data['potential_evaporation_sum'] /= 24
    hourly_data['total_precipitation_sum'] /= 24

    # Print the first few rows to verify the result
    print(hourly_data.head(48))  # Show two days of hourly data

    # Construct the path to the new pickle file
    new_pickle_file_path = current_dir / "data" / "Zwalm_data" / "preprocess_output" / "forcings_ERA5_david_E_EP.pkl"

    # Save the resulting DataFrame to the new pickle file
    hourly_data.to_pickle(new_pickle_file_path)

    print(f"Hourly data with correctly formatted Timestamps has been saved at {new_pickle_file_path}")
else:
    print(f"Error: The file {input_pickle_file_path} does not exist.")
```

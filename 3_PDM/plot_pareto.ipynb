```python
import numpy as np
import matplotlib.pyplot as plt

# Extracted from the LaTeX table (optimal values)
c_max = 375.66  # mm
c_min = 196.14  # mm
b = 0.523       # unitless

# Define the function f(c)
def f_c(c, c_min, c_max, b):
    norm = c_max - c_min
    return (b / norm) * ((c_max - c) / norm) ** (b - 1)

# Generate c values in the valid range
c = np.linspace(c_min, c_max, 500)
f_values = f_c(c, c_min, c_max, b)

# Plotting

plt.plot(c, f_values, color='black',linewidth = 5)

plt.savefig("paretoplot.png", dpi=1000, bbox_inches='tight') 
plt.show()
```
